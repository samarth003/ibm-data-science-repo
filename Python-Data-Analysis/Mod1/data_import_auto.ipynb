```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

df = pd.read_csv(url, header=None)
df.head(5)
```

```python
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-doors", "body-style", "drive-wheels",
           "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type", "num-cylinders",
           "engine-size", "fuel-system", "bore", "stroke", "compresson-ratio", "horsepower", "peak-rpm", "city-mpg",
           "highway-mpg", "price"]

df.columns = headers

df.head(5)
```

```python
path = "auto.csv"
auto_csv = df.to_csv(path_or_buf=path)
```
