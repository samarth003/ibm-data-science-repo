```python
from pycoingecko import CoinGeckoAPI

cg = CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_by_id(id= 'bitcoin', vs_currency= 'usd', days= 30)
```

```python
bitcoin_data
```

```python
import pandas as pd

df = pd.DataFrame(bitcoin_data['prices'], columns= ['Timestamp', 'Prices'])
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Timestamp'], unit= 'ms')
df
```

```python
#get candlestick plot
candlestick_data = df.groupby(df.Date.dt.date).agg({'Prices':['min', 'max', 'first', 'last']})
```

```python
candlestick_data
```

```python
import plotly.graph_objects as go
import plotly
try:
        fig = go.Figure(data=[go.Candlestick(x= candlestick_data.index, 
                                                        open= candlestick_data['Prices']['first'],
                                                        high= candlestick_data['Prices']['max'],
                                                        low= candlestick_data['Prices']['min'],
                                                        close= candlestick_data['Prices']['last']
                                                        )
                                            ]
                                    )
        fig.update_layout(xaxis_rangeslider_visible= False, xaxis_title= 'Date',
                       yaxis_title= 'Price(USD $)', title= 'Bitcoin Candlestick chart')

        plotly.offline.plot(fig, filename= './bitcoin_candlestick_graph.html', auto_open= True)
except Exception as e:
        print("Error observed ", e)
```
